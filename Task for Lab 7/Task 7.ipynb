# Task: Linear vs Polynomial Regression

## Objective
Build and compare Linear Regression and Polynomial Regression models then select the best performing model and interpret its learned equation.


## Dataset
Use the **Boston Housing dataset**.
    
    from sklearn.datasets import load_boston

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
# 1. Load the Dataset & 2. Convert to DataFrame
try:
    # Attempt to load from sklearn (works for sklearn versions < 1.2)
    from sklearn.datasets import load_boston
    boston = load_boston()
    df = pd.DataFrame(boston.data, columns=boston.feature_names)
    df['TARGET'] = boston.target
    print("Dataset loaded successfully from scikit-learn.")
    
except ImportError:
    # Alternative method: fetch directly from the source if sklearn import fails
    print("load_boston not found in sklearn. Loading from alternative source...")
    data_url = "http://lib.stat.cmu.edu/datasets/boston"
    raw_df = pd.read_csv(data_url, sep="\s+", skiprows=22, header=None)
    data = np.hstack([raw_df.values[::2, :], raw_df.values[1::2, :2]])
    target = raw_df.values[1::2, 2]
    
    # Feature names for the Boston dataset
    columns = ['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT']
    df = pd.DataFrame(data, columns=columns)
    df['TARGET'] = target
    print("Dataset loaded successfully from original URL.")
```

### 1. Data Preparation

```python
# 3. Data Inspection & Overview
print("\n--- Dataset Info ---")
df.info()
```

```python
print("\n--- Statistical Summary ---")
print(df.describe())
```

```python
print("\n--- Checking for missing values ---")
print(df.isnull().sum()) 
print("\nDataFrame Shape:", df.shape)
```

```python
# 4. Separate X and y
X = df.drop('TARGET', axis=1)
y = df['TARGET']

# 5. Train/Test Split (80% training, 20% testing)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=101)
```

```python
# 6. Feature Scaling
scaler = StandardScaler()
# Fit on training data ONLY, then transform both
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

print("\nData Preparation Complete! X_train_scaled and X_test_scaled are ready.")
```

### 2. Linear Regression Model
- Train a Linear Regression model
- Make predictions on the test set
- Compute evaluation metrics:
  - MAE
  - MSE
  - RMSE

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np

# 1. Initialize the Linear Regression model
lr_model = LinearRegression()
```

```python
# 2. Train the model using the scaled training data
lr_model.fit(X_train_scaled, y_train)

# 3. Make predictions on the scaled test data
y_pred_lr = lr_model.predict(X_test_scaled)
```

```python
# 4. Compute evaluation metrics
mae_lr = mean_absolute_error(y_test, y_pred_lr)
mse_lr = mean_squared_error(y_test, y_pred_lr)
rmse_lr = np.sqrt(mse_lr)

```

```python
# Print the results
print("--- Linear Regression Model Evaluation ---")
print(f"Mean Absolute Error (MAE): {mae_lr:.4f}")
print(f"Mean Squared Error (MSE): {mse_lr:.4f}")
print(f"Root Mean Squared Error (RMSE): {rmse_lr:.4f}")
```

### 3. Polynomial Regression Model
- Train a Polynomial Regression model
- Make predictions on the test set
- Compute evaluation metrics:
  - MAE
  - MSE
  - RMSE


```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np

degrees = [2, 3, 4]
best_degree = None
best_rmse = float('inf')
best_poly_metrics = {}
```

```python
print("--- Testing Different Polynomial Degrees ---")

for d in degrees:
    # 1. Transform features into polynomial features for degree 'd'
    poly = PolynomialFeatures(degree=d)
    X_train_poly = poly.fit_transform(X_train_scaled)
    X_test_poly = poly.transform(X_test_scaled)
    
    # 2. Initialize and train the model
    poly_model = LinearRegression()
    poly_model.fit(X_train_poly, y_train)
    
    # 3. Make predictions
    y_pred_poly = poly_model.predict(X_test_poly)
    
    # 4. Compute metrics
    mae = mean_absolute_error(y_test, y_pred_poly)
    mse = mean_squared_error(y_test, y_pred_poly)
    rmse = np.sqrt(mse)
    
    print(f"Degree {d} -> MAE: {mae:.4f} | MSE: {mse:.4f} | RMSE: {rmse:.4f}")
    
    # 5. Track the best degree based on the lowest RMSE
    if rmse < best_rmse:
        best_rmse = rmse
        best_degree = d
        best_poly_metrics = {'MAE': mae, 'MSE': mse, 'RMSE': rmse}
```

```python
print(f"\n>>> The best Polynomial Degree based on test RMSE is: {best_degree} <<<")
```

### 4. Model Comparison
- Compare Linear vs Polynomial Regression results
- Identify which model performs better on the test data and justify your decision


```python
import pandas as pd

# Creating a summary dataframe using your actual test results
results = {
    'Model': ['Linear Regression', 'Polynomial Regression (Degree 2)'],
    'MAE': [3.9994, 2.8468],
    'MSE': [33.2689, 17.1358],
    'RMSE': [5.7679, 4.1395]
}

comparison_df = pd.DataFrame(results)

print("--- Final Model Comparison ---")
print(comparison_df.to_string(index=False))
```

The Polynomial Regression model (Degree 2) is the clear winner and performs significantly better on the unseen test data.

Justification:

Lower Average Error: The Mean Absolute Error (MAE) dropped from ~4.00 in the linear model to ~2.85. In the context of the dataset, this means our average prediction error improved from roughly $4,000 off the actual price to only $2,850 off.

Fewer Extreme Mistakes: The Root Mean Squared Error (RMSE) also saw a substantial decrease (from 5.77 down to 4.14). Because RMSE heavily penalizes large errors, this drop proves that the Degree 2 model makes far fewer massive miscalculations than the standard linear model.

Capturing Complexity: The standard Linear Regression model assumes a perfectly straight-line relationship between every feature (like crime rate or age) and the house price. By expanding our features to a 2nd-degree polynomial, the model was able to capture the natural curves and interactions present in the real-world housing market, leading to a much tighter fit without overfitting (which we observed happening at Degrees 3 and 4).
